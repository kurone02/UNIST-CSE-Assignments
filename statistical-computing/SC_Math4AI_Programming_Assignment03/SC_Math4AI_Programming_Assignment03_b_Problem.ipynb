## Problem 5-2. log-Likelihood($\beta$)

* Logistic Regression Problem:<br/>
$y_i \sim Ber(\pi(x_i))$, $\pi(x_i) = \dfrac{exp(\beta_0+\beta_1x_i)}{1+exp(\beta_0+\beta_1x_i)}$
<br/><br/>
Hence, $y_i$ is a binary variable taking values 0 or 1 with $P(y_i=1|x_i) = \pi(x_i), P(y_i=0|x_i) = 1-\pi(x_i)$<br/> and probability mass function of $y_i$ is <br/>
$p(y_i|x_i) = \pi(x_i)^{y_i} \times (1-\pi(x_i))^{1-y_i}$
<br/>
Suppose that $y_1, y_2, ... , y_N$ are independent. Then the joint probability mass function of $y_1, y_2, ... , y_N$ is <br/>
$p(y_1, ... , y_N|x_1, ... , x_N) = \prod_{i=1}^{N} \pi(x_i)^{y_i} \times (1-\pi(x_i))^{1-y_i}$.<br/>
==> $Likelihood(\beta)=L(\beta)= \prod_{i=1}^{N} \pi(x_i)^{y_i} \times (1-\pi(x_i))^{1-y_i}$<br/>
==> $Log-likelihood(\beta) = logL(\beta) = l(\beta)$
